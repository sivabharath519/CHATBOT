# STEPS

# Clean the data
- convert into lower case
- remove numbers
- remove special characters(,.'<>..)
- > Take root words.(Remove all tenses and plurals)
- Tokenization --> Taking words

    - stemming (We use now this model) ----> Prcoess of checking the root node
    
    
           OR
           
          
    - lemmatiztaion - normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemm
 
 
 - remove stop words
 - spell correction


```python
import nltk                       # Natrual Language Tool Kit
from nltk import word_tokenize    # Word Tokenization ---> It makes tokens(individual words)
from nltk.stem import wordnet     # To perform Lemmitization
from nltk import pos_tag          # For parts f speech
```

```python
# ----------------------------Text Normalization-----------------------------------#

def lemmatiztaion(text):
    text = str(text).lower()  # Converted into lower case
    
    spl_char_text = re.sub(r'[^a-z]',' ',text) # removed numbers and special characters
    
    tokens = word_tokenize(spl_char_text)   # word tokenization
    
    lemma = wordnet.WordNetLemmatizer()    # initalize lemmatizer
    
    tags_list = pos_tag(tokens, tagset= None) # Parts of speech
    
    # Created empty list
    
    lemma_words= []
    
    for token, pos_token in tags_list:
        
        if pos_token.startswith('V'):     # verb
            pos_val = 'v'
        elif pos_token.startswith('J'):   # adjective
            pos_val = 'a'
        elif pos_token.startswith('R'):   # adverb
            pos_val = 'r'
        else:    # any parts of speech except verb, adjective, adverb
            pos_val ='n'
            
        lemma_token = lemma.lemmatize(token, pos_val)   # computing
        lemma_words.append(lemma_token) # append values in list
        
            
    return " ".join(lemma_words) 
        
    
    
```

# Cosine Similarity
- Cosine similarity is a metric used to measure how similar the documents are irrespective of their size.
- A commonly used approach to match similar documents is based on counting the maximum number of common words between the documents.
- Mathematically, it measures the cosine of the angle between two vectors projected in a multi-dimensional space.
- The cosine similarity is advantageous because even if the two similar documents are far apart by the Euclidean distance (due to the size of the document), chances are they may still be oriented closer together.


**NOTE :**
###  The smaller the angle, higher the cosine similarity

```python
from sklearn.metrics import pairwise_distances
```

```python
cosine_simmi = 1 - pairwise_distances(df_bow, Question_bow, metric = 'cosine')
```

```python
print(cosine_simmi)
```

```python
index_value = cosine_simmi.argmax()
```

```python
df['Text Response'].loc[index_value]
```

### Drawback of Bag_OF_Words    (BOW)
- Problem of Bag of words is stopwords
- stop words are common repeated words
- Bag of words gives importance for each word in the document

# TF-IDF


> - #### Term frequency- Inverse Document Frequency
- How many times word repeated in the document
- _ With the help of TF-IDF we can reduce the stop words


### DIFFerence between Bag_of_word VS TF-IDF
> > **Bag of words gives importance for each word in the document, but TF-IDF gives based on their weights**

```python
from sklearn.feature_extraction.text import TfidfVectorizer   
```

```python
tfidf = TfidfVectorizer()  # initiallizing Tfidf
```

```python
trans_tfidf = tfidf.fit_transform(df['lemmatize_text']).toarray()

# transforing data into ARRAY
```

```python
# returns all unique words of data witha score of that word


data_tfidf = pd.DataFrame(trans_tfidf, columns= tfidf.get_feature_names())
data_tfidf.head(3)
```

### Now find Cosine Simmilarity

```python
Question_tfidf = tfidf.transform([Question_lemma]).toarray()

      # applied tfidf
```

```python
cosine = 1 - pairwise_distances(data_tfidf, Question_tfidf, metric = 'cosine')   # applying cosine similarity
```

```python
cosine
```

```python
index_value_new = cosine.argmax()
```

```python
index_value_new
```

```python
 # define a function that  response to their query
    

def chatter(text):
    lemma = lemmatiztaion(text) # calling the funtion 
    tf = tfidf.transform([lemma]).toarray()
    cosine = 1 - pairwise_distances(data_tfidf, tf, metric = 'cosine')   # applying cosine similarity
    index_value_new = cosine.argmax()
    return df['Text Response'].loc[index_value_new]
```
